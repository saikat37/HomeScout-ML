```python
df=df[df['price'] != 'Price on Request']
df.sample()
```

```python
def lac_crore (s):
    if type(s)==float:
        return s
    else:
        if s[1]=="Lac":
            return round(float(s[0])/100,2)
        else:
            return round(float(s[0]),2)
```

```python
df["price"]=df['price'].str.split(" ").apply(lac_crore)
```

```python
df['price_per_sqft'] = df['price_per_sqft'].str.split('/').str.get(0).str.replace('₹','').str.replace(',','').str.strip().astype('float')
```
